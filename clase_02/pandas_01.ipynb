Muchas veces, en un Dataframe se necesita realizar operaciones entre  la fila actual y la fila anterior, lo cual puede ser complejo si no se utilizan las funciones correctas. A continuaci칩n se trabajan algunas de estas funciones:

1. **`shift()`**: Se utiliza para mover hacia arriba o hacia abajo los valores de una columna o serie de datos.
2. **`cumsum()`**: es una funci칩n que se utiliza para calcular la suma acumulativa de valores a lo largo de un eje en un DataFrame o una Serie.
3. **`pct_change()`**: es una funci칩n que se utiliza para calcular el cambio porcentual entre los elementos de una serie o columna en un DataFrame. 
4. **`rank()`**: es una funci칩n que se utiliza para asignar un rango a los elementos de una serie o columna en un DataFrame.

```python
# aplicar funciones
df['shift'] = df['duration'].shift() # se muestra el valor de la fila anterior (la primera fila en este caso es NaN)
df['cumsum'] = df['duration'].cumsum()  # suma acumulada entre la fila actual y todas las anteriores
df['pct_change'] = df['duration'].pct_change() # porcentaje de cambio entre la fila actual y la anterior 
df['rank'] = df['duration'].rank() # ranking de los valores (donde 1 es el menor valor)
df.head()
```

```python
cols = ['shift', 'cumsum','pct_change', 'rank']
df = df.drop(cols,axis=1)
df.head()
```

### Filtrar Datos

Para filtrar datos en Pandas, se utiliza el m칠todo `loc()` o `iloc()`, dependiendo de si queremos filtrar por etiquetas de 칤ndice o por posici칩n. Para efectos pr치cticos, utilizaremos solo `loc()`.

```python
# 'fecha' mayor a 2000
df_new = df.loc[df['year_start'] >= 2000]
df_new.head()
```

```python
# crear condicion
valor_objetivo = 2000 
condicion = (df['year_start'] >= valor_objetivo)
condicion.head()
```

```python
# aplicar condicion
df_new = df.loc[condicion]
df_new.head()
```

Veamos distintos tipos de filtros:

```python
# mayor o igual a 2000
df_new = df.loc[df['year_start'] >= 2000]
df_new.head()
```

```python
# entre 2005-2015
df_new = df.loc[df['year_start'].between(2005,2015)]
df_new.head()
```

```python
# solo 2000
df_new = df.loc[df['year_start']==2000]
df_new.head()
```

```python
# varias condiciones condiciones
df_new = df.loc[(df['year_start']==2000) & (df['duration']>5)]
df_new.head()
```

```python
# Filtrar por patr칩n de texto 
df_new = df.loc[df['name'].str.contains('Michael')]
df_new.head()
```

### Valores Nulos y Duplicados

#### Valores Nulos

Un valor nulo (o faltante) representa la ausencia de un valor en una celda espec칤fica de un DataFrame o una Serie. Los valores nulos pueden ocurrir por varias razones, como datos perdidos o no disponibles, errores de medici칩n o problemas de entrada de datos.

Los valores nulos se representan en Pandas mediante el objeto `NaN` (acr칩nimo de "Not a Number"). `NaN` es un valor especial de punto flotante definido en el est치ndar **IEEE** para representar valores no definidos o indefinidos. En Pandas, los valores nulos se representan como `NaN` para las Series y DataFrames que utilizan datos num칠ricos, y como `None` para las Series y DataFrames que utilizan datos no num칠ricos.

```python
# valores nulos
df.isnull().sum()
```

```python
# Eliminar las filas que contienen valores nulos
df_new = df.dropna()
df_new.head()
```

```python
# Rellenar los valores nulos con un valor espec칤fico, por ejemplo cero
df_new = df.fillna(0)
df_new.head()
```

* El m칠todo `duplicated()` devuelve un booleano que indica si una fila es duplicada o no, es decir, si existe otra fila con los mismos valores.

### Manipulaci칩n de Fechas

Pandas se desarroll칩 en el contexto del modelado financiero, por lo que, contiene varias herramientas para trabajar con fechas, horas y datos indexados por tiempo.

Comenzaremos por entendender las herramientas para manejar fechas y horas en Python, antes de pasar m치s espec칤ficamente a las herramientas proporcionadas por Pandas. 

```python
import datetime
now = datetime.datetime.now()
print(now)
```

Puedes acceder a partes espec칤ficas de un objeto `datetime.datetime`, como el a침o, el mes, el d칤a, la hora, el minuto y el segundo. Algunos de los atributos m치s comunes son:

* `year`: representa el a침o de la fecha y la hora.
* `month`: representa el mes de la fecha y la hora, como un n칰mero entre 1 y 12.
* `day`: representa el d칤a del mes de la fecha y la hora, como un n칰mero entre 1 y 31.
* `hour`: representa la hora del d칤a de la fecha y la hora, como un n칰mero entre 0 y 23.
* `minute`: representa los minutos de la hora de la fecha y la hora, como un n칰mero entre 0 y 59.
* `second`: representa los segundos de la hora de la fecha y la hora, como un n칰mero entre 0 y 59.
* `microsecond`: representa los microsegundos de la hora de la fecha y la hora, como un n칰mero entre 0 y 999999.

Veamos todo lo anterior, aplicado al dataframe que estamos trabajando:

```python
# cambiar formato de la fecha 
df['new_birth_date'] = pd.to_datetime(df['birth_date'])
df.head()
```

```python
# ver atributos
df.dtypes
```

```python
# calcular atributos como a침o, mes, dia, ...
df['year'] = df['new_birth_date'].dt.year
df['month'] = df['new_birth_date'].dt.month
df['day'] = df['new_birth_date'].dt.day
df['hour'] = df['new_birth_date'].dt.hour
df['minute'] = df['new_birth_date'].dt.minute
df['second'] = df['new_birth_date'].dt.second
df['microsecond'] = df['new_birth_date'].dt.microsecond

df.head()
```

```python
# lista de columnas a eliminar
columnas_a_eliminar = ['year', 'month', 'day', 'hour', 'minute', 'second', 'microsecond']

# eliminar las columnas
df = df.drop(columnas_a_eliminar, axis=1)

# mostrar dataframe
df.head()
```
